```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('./archive/StudentsPerformance.csv')
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
sns.pairplot(df[['math score','reading score','writing score']], height=2)
```

```python
def avg_score(dt):
    return (dt['math score']+dt['reading score']+dt['writing score'])/3
df['avg_score']=df.apply(avg_score,axis=1)
df.head()
```

```python
class_score=['math score','reading score','writing score','avg_score']
for i in class_score:
    sns.catplot(x='lunch',y=i,hue='gender',kind='boxen',data=df,height=3,palette=sns.color_palette(['red','blue']))
    plt.title(i)
```

```python
sns.catplot(x='test preparation course',y='avg_score',hue='gender',kind='boxen',data=df,height=3,palette=sns.color_palette(['red','blue']))
plt.title('avg_score')
```

```python
sns.catplot(x='parental level of education',y='avg_score',hue='gender',kind='boxen',data=df,height=10)
plt.title('avg_score')
plt.legend(loc='lower')
```
